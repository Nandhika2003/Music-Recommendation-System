```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('dataset.csv')
df.head(3)
```

```python
df.shape
```

```python
df = df.sample(n = 5000).drop('link',axis = 1).reset_index(drop=True)
```

```python
#Cleaning
df['text'][0]
```

```python
df['text'] = df['text'].str.lower().replace(r'[^\w\s]','').replace(r'\n',' ',regex=True)
```

```python
df['text'][0]
```

```python
#Tokenization and Steming
import nltk
from nltk.stem.porter import PorterStemmer

ps = PorterStemmer()

#stemming = [] (more memory error)
def tokenization(txt):
    tokens = nltk.word_tokenize(txt)
    #for w in tokens:
        #stemming.append(ps.stem(w))
    stemming = [ps.stem(w) for w in tokens]
        
    return " ".join(stemming)
```

```python
tokenization('this is loving')
```

```python
df['text'] = df['text'].apply(lambda x: tokenization(x))
```

```python
df['text']
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer # Tf = W/TW, Idf = TD/WC, Formula = Tf*Idf
from sklearn.metrics.pairwise import cosine_similarity #(theta = 0(similar) and 90(not similar)
```

```python
tfid = TfidfVectorizer(stop_words='english')  # remove the words with no meaning - this, is, to , form
matrix = tfid.fit_transform(df['text'])
```

```python
matrix.shape
```

```python
similarity = cosine_similarity(matrix)
```

```python
def recommendation(song):
    idx = df[df['song'] == song].index[0]
    distances = sorted(list(enumerate(similarity[idx])),reverse=False, key=lambda x:x[1])
    
    songs = []
    for i in distances[1:21]:
        songs.append(df.iloc[i[0]].song)
    return songs
```

```python
recommendation('Dangerous Tonight')
```

```python
recommendation('Bird Of Pray')
```
